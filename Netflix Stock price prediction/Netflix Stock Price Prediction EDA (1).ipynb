# Features in the Netflix Stock Price Prediction Dataset

## Key Features Explained

### 1. **High**
   - **Definition:** The highest stock price recorded during the trading day.
   - **Why it's useful:** The **High** price indicates the peak price at which the stock traded during the day. It provides insights into the maximum demand for the stock in a given period. Understanding the high price helps in identifying potential price resistance levels, which are useful for predicting price reversals or momentum.
   
### 2. **Low**
   - **Definition:** The lowest stock price recorded during the trading day.
   - **Why it's useful:** The **Low** price reflects the minimum price at which the stock traded. It provides insights into the lowest demand or the support level for the stock. Identifying the low price is critical for understanding market trends, detecting potential price bottoms, and anticipating price recoveries.

### 3. **Open**
   - **Definition:** The stock price at the beginning of the trading day.
   - **Why it's useful:** The **Open** price is the starting point of the day’s trading and often reflects market sentiment when the market opens. Analyzing the opening price can reveal market sentiment, as significant price changes from the previous day's close to the open indicate news or market events.

### 4. **Close**
   - **Definition:** The stock price at the end of the trading day.
   - **Why it's useful:** The **Close** price is the most widely used price by investors, as it reflects the stock’s value at the end of the trading day. It is critical in technical analysis, and it's used to calculate various technical indicators such as moving averages and relative strength index (RSI). The close price provides a snapshot of the stock's market value at the end of the session.

### 5. **Adj Close (Adjusted Close)**
   - **Definition:** The adjusted closing price, accounting for factors like stock splits, dividends, and other adjustments.
   - **Why it's useful:** The **Adj Close** price is important for long-term investors because it adjusts for corporate actions that affect the stock’s value, such as dividends or stock splits. This feature is essential for accurately tracking the performance of a stock over time, especially for historical analysis.

### 6. **Volume**
   - **Definition:** The total number of shares traded during the trading day.
   - **Why it's useful:** **Volume** is a key indicator of market activity and liquidity. High trading volume can indicate strong investor interest, while low volume can suggest the opposite. Volume often correlates with price changes, as it provides insights into how much the market is willing to support or resist price movements. High volume combined with price increases may indicate a strong trend.

---

## Why These Features are Useful for Stock Price Prediction  

- **Price Trends and Patterns:** Features like **High**, **Low**, **Open**, and **Close** represent the core price movements of the stock, which are essential for detecting trends, patterns, and market cycles. These price points help identify support and resistance levels, price breakouts, and reversal patterns.
  
- **Market Sentiment and Demand:** The **High** and **Low** prices give insights into market sentiment and the volatility of the stock on a given day. By analyzing how the stock behaves between its high and low points, we can gain clues about future price movements.

- **Investor Behavior:** **Volume** data helps in understanding how many investors are actively trading the stock. A surge in volume often precedes significant price movements, providing an early indicator of market shifts.

- **Model Accuracy:** These features are the foundational inputs for machine learning models like Random Forests and XGBoost, which use them to capture the relationships between past prices and future movements. They help the model make accurate predictions by understanding how stock prices are influenced by past performance.

---

## Conclusion  
These features are crucial for building accurate stock price prediction models, as they provide a comprehensive view of market behavior and stock performance. By analyzing them in combination, machine learning algorithms can effectively predict future stock prices and help investors make data-driven decisions.

# Detecting the outliers 

```python
# Checking for outliers in key numerical columns using the Interquartile Range (IQR) method

# Define a function to detect outliers in a column using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers[[column]]

# Columns to check for outliers
columns_to_check = ['Open', 'High', 'Close', 'Adj Close', 'Low','Volume']


# Detecting outliers in each column
outliers_dict = {col: detect_outliers(df, col) for col in columns_to_check}
outliers_dict_summary = {col: outliers_dict[col].shape[0] for col in columns_to_check}

outliers_dict_summary
```
